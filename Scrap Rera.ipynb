```python
def getAllDetailsbyProject(projID):
    alldatabyprojectid = f"http://3.111.75.239/project_reg/public/alldatabyprojectid/{projID}"
    getproject_details = f"http://3.111.75.239/project_reg/public/getproject-details/{projID}"
    getproject_banks = f"http://3.111.75.239/project_reg/public/project-app/getproject-banks/{projID}"
    getProjectProgressReports_ByPrjRegId = f"http://3.111.75.239/project_reg/public/getProjectProgressReports-ByPrjRegId/{projID}"
    getproject_doc = f"http://3.111.75.239/project_reg/public/getproject-doc/{projID}"
    getProjectLocations = f"http://3.111.75.239/maplocation/public/getProjectLocations/{projID}"
    projectDetails = [alldatabyprojectid,getproject_details,getproject_banks,getProjectProgressReports_ByPrjRegId,getproject_doc,getProjectLocations]
    dataDictProject = {}
    for dataPoint in tqdm(projectDetails):
        jsonData = requests.get(dataPoint).json()
        # dataDf = pd.DataFrame.from_dict(jsonData['data'])
        dataDictProject[dataPoint.split('/')[-2]] = jsonData
        time.sleep(2)

    time.sleep(5)
    form3ID = dataDictProject['alldatabyprojectid']['data']['formThreeId']
    form2ID = dataDictProject['alldatabyprojectid']['data']['formTwoId']
    form1ID = dataDictProject['alldatabyprojectid']['data']['formOneId']

    getform_three_byid = f"http://3.111.75.239/formthree/public/getform-three-byid/{form3ID}"

    getform_two_by_id = f"http://3.111.75.239/formtwo/public/getform-two-by-id/{form2ID}"

    getfrom_one_byformone_id = f"http://3.111.75.239/formone/public/getfrom-one-byformone-id/{form1ID}"

    extraDetails = [getform_three_byid,getform_two_by_id,getfrom_one_byformone_id]
    dataDictExtra = {}
    for dataPoint in tqdm(extraDetails):
        jsonData = requests.get(dataPoint).json()
        dataDictExtra[dataPoint.split('/')[-2]] = jsonData
        time.sleep(2)
    # print(dataDict)
    jsonDict = {'projectDetails':dataDictProject,
            'projectExtra':dataDictExtra}

    with open(f"Saved Jsons/{projID}_{form3ID}.json", "w") as outfile:
        json.dump(jsonDict, outfile)
    return jsonDict



```
